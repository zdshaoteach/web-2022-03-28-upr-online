## for loops

`for` loops are more widely used [reference: w3schools](https://www.w3schools.com/python/python_for_loops.asp) than [`while` loops](https://www.w3schools.com/python/python_while_loops.asp)

why are we using `for` loops?
- sequence data types (list, tuple, string, etc.) are iterative
- repeat the same operation for different elements

```python
# repeat on each item from a list
shopping_list = ['rice', 'olive oil', 'salt', 'sugar', 'apple']
for item in shopping_list:
    print('I plan to buy', item)
```

## python conditions



- the `if ... else ...` statement [reference: w3schools](https://www.w3schools.com/python/python_conditions.asp)

```python
########################################
# exercise: count the bill
shopping_list = [
    ('rice', 2.0), 
    ('olive oil', 9.5), 
    ('salt', 0.9), 
    ('sugar', 1.5), 
    ('apple', 2.9)
]
# how to write a loop to count the total payment?
total_amount = 0
for index, item in enumerate(shopping_list):
    if item[1] > 5.0:
        print('I do not want to buy the #{} item {} \n \
            because it is too expensive'.format(
            index, item[0])
        )
    else:
        print("The #{} item I want to buy is {}, \n \
            and it costs {} dollars.".format(
            index, item[0], item[1])
        )
        total_amount += item[1]
print('total amount of the bill is', total_amount, 'dollars.')
```

```python
# what if I have a long shopping list but I do not have enough money?
########################################
# exercise: count the bill
# a while loop
shopping_list = [
    ('rice', 2.0), 
    ('olive oil', 3.5), 
    ('salt', 0.9), 
    ('sugar', 1.5), 
    ('apple', 2.9)
]
wallet = 6.0
# initialize loop variables
total_amount = 0.0
index = 0
payment = shopping_list[index][1]
while payment < wallet:
    total_amount = payment
    print("The #{} item I want to buy is {}, and it costs {} dollars.".format(
            index, shopping_list[index][0], shopping_list[index][1]))
    index += 1
    payment = total_amount + shopping_list[index][1]  
print('total amount of the bill is', total_amount, 'dollars.')
# do you have a program similar to this?
# is it functioning as expected?
```

```python
########################################
# exercise: find the maximum GDP per capita from each csv file
import glob
import pandas as pd

def find_maximum_gdp_per_capita(filename, target_col='gdpPercap_2007'):
    data = pd.read_csv(filename, index_col='country')
    #print(data.head(5))
    target_col
    if target_col in list(data):
        result = data[target_col].max()
        return data.loc[data[target_col] == result, target_col]
    else:
        print(target_col, 'does not exist in csv')

for file_name in glob.glob('data/*'):
    maximum_gdp = find_maximum_gdp_per_capita(
        target_col='gdpPercap_2002', filename=file_name)
    print(file_name, 'maximum gdp per capita is', maximum_gdp)
```

## variable scope

- global variables
- module variables
- local variables: inside a function

```python
def set_age(value):
    #global age
    age = value
    return age

age = 72

print('set age by function', set_age(30))
print('the value of "global variable age" is', age)
```

```python
def set_age(value):
    global age
    age = value
    return age

age = 72

print('set age by function', set_age(30))
print('the value of "global variable age" is', age)
```

```python
########################################
# exercise: write a function that finds all (positive) prime numbers 
# that are less than a given integer
```

```python
def find_prime(value):
    assert isinstance(value, int)
    prime_numbers = [2]
    
    if value < 2:
        return []
    elif value == 2:
        return prime_numbers
    
    for num in range(3,value+1):
        is_prime = True
        for div in prime_numbers:
            if num % div == 0:
                is_prime = False
                
        if is_prime:
            prime_numbers.append(num)
            
    return prime_numbers

assert find_prime(1) == []
assert find_prime(2) == [2]
assert find_prime(3) == [2,3]
assert find_prime(4) == [2,3]
assert find_prime(5) == [2,3,5]
assert find_prime(6) == [2,3,5]
assert find_prime(7) == [2,3,5,7]
assert find_prime(8) == [2,3,5,7]
assert find_prime(9) == [2,3,5,7]
assert find_prime(10) == [2,3,5,7]
assert find_prime(11) == [2,3,5,7,11]
assert find_prime(12) == [2,3,5,7,11]
assert find_prime(13) == [2,3,5,7,11,13]
assert find_prime(14) == [2,3,5,7,11,13]
assert find_prime(15) == [2,3,5,7,11,13]
```
