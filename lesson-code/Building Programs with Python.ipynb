## visualize data with matplotlib

[basic use of matplotlib](https://matplotlib.org/stable/tutorials/introductory/usage.html#sphx-glr-tutorials-introductory-usage-py)

```python
# create a plot
import matplotlib.pyplot as plt
# reference of customization :
# https://matplotlib.org/stable/tutorials/introductory/customizing.html
plt.rcParams["figure.figsize"] = (16,9)
plt.rcParams['figure.dpi'] = 300

country = selected.index
population = selected.pop_2007
population = population / 1000000

plt.plot(country, population)
plt.xlabel('country')
plt.ylabel('population (million)')
#plt.xticks(rotation=0)
plt.show()
```

```python
# quick plot with pandas dataframe
plot_data = selected.loc[:, 'pop_1997':'pop_2007']
plot_data.plot()
plt.show()
```

```python
# another way to generate a plot
fig, ax = plt.subplots()  # a figure with a single Axes
ax.plot(country, population)
ax.set_xlabel('country')
ax.set_ylabel('population (million)')
plt.show()
```

```python
# generate subplots
fig, axs = plt.subplots(1, 2) # a figure with a 1x2 grid of Axes

axs[0].plot(selected.index, selected.pop_2002/1000000)
axs[0].set_xlabel('country')
axs[0].set_ylabel('population (million)')
axs[0].set_title('2002')
axs[0].set_ylim(0, 18)

axs[1].plot(selected.index, selected.pop_2007/1000000)
axs[1].set_xlabel('country')
axs[1].set_ylabel('population (million)')
axs[1].set_title('2007')
axs[1].set_ylim(0, 18)

fig.autofmt_xdate(rotation=45)
fig.align_ylabels(axs)
plt.savefig('plot-2002-2007.png')
#plt.show()
```

```python
########################################
# exercise: generate subplots
fig, axs = plt.subplots(2, 2) # a figure with a 2x2 grid of Axes

axs[0, 0].plot(selected.index, selected.pop_1992/1000000)
axs[0, 0].set_xlabel('country')
axs[0, 0].set_ylabel('population (million)')
axs[0, 0].set_title('1992')
axs[0, 0].set_ylim(0, 18)

axs[0, 1].plot(selected.index, selected.pop_1997/1000000)
axs[0, 1].set_xlabel('country')
axs[0, 1].set_ylabel('population (million)')
axs[0, 1].set_title('1997')
axs[0, 1].set_ylim(0, 18)

axs[1, 0].plot(selected.index, selected.pop_2002/1000000)
axs[1, 0].set_xlabel('country')
axs[1, 0].set_ylabel('population (million)')
axs[1, 0].set_title('2002')
axs[1, 0].set_ylim(0, 18)

axs[1, 1].plot(selected.index, selected.pop_2007/1000000)
axs[1, 1].set_xlabel('country')
axs[1, 1].set_ylabel('population (million)')
axs[1, 1].set_title('2007')
axs[1, 1].set_ylim(0, 18)

fig.autofmt_xdate(rotation=45)
fig.align_ylabels(axs)
plt.savefig('plot-1992-2007.png')
#plt.show()
```
